I. Write a Python program to plot the values of the following functions, which represent various
time complexity classes:
1. Constant:             
2. Logarithmic:           
3. Linear:               
4. Linearithmic:          
5. Quadratic:             
6. Cubic:               
7. Exponential:           
8. Factorial:             
f(n) = 1
f(n) = log₂(n)
f(n) = n
f(n) = n × log₂(n)
f(n) = n²
f(n) = n³
f(n) = 2ⁿ
f(n) = n!
For better visualization, choose n from 1 to 20. Use logarithmic scaling or limit the y-axis range
where appropriate to accommodate fast-growing functions like 2ⁿ and n!.

```
 import math
 import numpy as np
 import matplotlib.pyplot as plt
 n = np.arange(1, 21)
 constant = np.ones_like(n)
 logarithmic = np.log2(n)
 linear = n
 linearithmic = n * np.log2(n)
 quadratic = n ** 2
 cubic = n ** 3
 exponential = 2 ** n
 factorial = np.array([math.factorial(i) for i in n])
 plt.figure(figsize=(12, 8))
 plt.plot(n, constant, label='Constant (1)')
 plt.plot(n, logarithmic, label='Logarithmic (log2 n)')
 plt.plot(n, linear, label='Linear (n)')
 plt.plot(n, linearithmic, label='Linearithmic (n log2 n)')
 plt.plot(n, quadratic, label='Quadratic (n^2)')
 plt.plot(n, cubic, label='Cubic (n^3)')
 plt.plot(n, exponential, label='Exponential (2^n)')
 plt.plot(n, factorial, label='Factorial (n!)')
 plt.yscale('log')
 plt.xlabel('n')
 plt.ylabel('f(n)')
 plt.title('Time Complexity Functions')
 plt.legend()
 plt.grid(True, which="both", ls="--")
 plt.show()
```

. A teacher is preparing the final merit list for a class test. Each student has a name and a
score out of 100. The list must be sorted so that:
• Students are arranged in descending order of marks.
• If two students have the same mark, they should be sorted alphabetically by name.
Write a program that:
1. Accepts input for n students: name (string) and marks (integer).
2. Sorts the student records using a custom implementation of a sorting algorithm
3. Displays the sorted list in a clean, tabular format showing rank, name, and marks.
4. Tracks and displays:
o Total number of comparisons
o Total number of swaps

```
def bubble_sort_students(students):
    n = len(students)
    comparisons = 0
    swaps = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            # Sort primarily by marks (descending), then name (ascending)
            mark_j, name_j = students[j][1], students[j][0]
            mark_next, name_next = students[j+1][1], students[j+1][0]

            if (mark_j < mark_next) or (mark_j == mark_next and name_j > name_next):
                # Swap if out of order
                students[j], students[j+1] = students[j+1], students[j]
                swaps += 1
    return comparisons, swaps
def main():
    students = []
    n = int(input("Enter the number of students: "))
    for _ in range(n):
        name = input("Enter student name: ").strip()
        marks = int(input("Enter marks (0-100): "))
        students.append([name, marks])
    comparisons, swaps = bubble_sort_students(students)
    # Display sorted list
    print("\nFinal Merit List:")
    print("{:<5}{:<15}{:<10}".format("Rank", "Name", "Marks"))
    print("-"*32)
    for idx, (name, marks) in enumerate(students, start=1):
        print("{:<5}{:<15}{:<10}".format(idx, name, marks))
    print("\nTotal Comparisons:", comparisons)
    print("Total Swaps:", swaps)
if __name__ == "__main__":
    main()
```

III. You are developing a module for a retail store’s inventory management system. The system
stores a list of product IDs (integers) and needs to quickly locate a product when a customer
requests it.
1. Accept a list of n product IDs (integers) as input.
2. Accept a product ID to search (target).
3. Implement the search algorithms
4. For each algorithm:
o Return whether the product is found and at what index.
o Count and display the number of comparisons made.
o Measure the time taken to perform the search.

In an actual software development scenario, especially for a retail store's inventory management system, it's crucial to consider that product IDs might not always be sorted. Therefore, both linear and binary search algorithms have their place. A linear search would be applicable for unsorted lists, while a binary search would be highly efficient for sorted lists. The choice of algorithm would depend on whether the product ID list is guaranteed to be sorted or if sorting is feasible as a preprocessing step.

Let's outline the plan for developing the module:

Input Collection:

Prompt the user to enter the number of product IDs (n).

Prompt the user to enter n product IDs, ensuring they are integers.

Prompt the user to enter the target product ID to search for.

Algorithm Implementations:

Linear Search:

Iterate through the list from the beginning to the end.

Compare each element with the target.

Increment a comparison counter for each comparison.

If the target is found, return True and its index.

If the end of the list is reached without finding the target, return False.

Binary Search (requires a sorted list):

First, ensure the input list of product IDs is sorted. This is a critical prerequisite for binary search. We'll sort it if it's not already.

Initialize low to 0 and high to n-1.

While low is less than or equal to high:

Calculate mid = (low + high) // 2.

Increment a comparison counter.

If list[mid] is equal to target, return True and mid.

If list[mid] is less than target, set low = mid + 1.

If list[mid] is greater than target, set high = mid - 1.

If the loop finishes, return False.

Performance Measurement and Display:

For each algorithm:

Record the start time before executing the search.

Execute the search algorithm.

Record the end time after the search.

Calculate the time taken (end_time - start_time).

Display:

Whether the product was found and its index (if found).

The total number of comparisons made.

The time taken for the search (e.g., in milliseconds or microseconds).

This comprehensive approach will demonstrate the functionality and performance characteristics of both search algorithms, which is crucial for making informed decisions in an inventory management system.

```
import time
def linear_search(arr, target):# Linear Search Algorithm
    comparisons = 0
    start_time = time.time()
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            end_time = time.time()
            return True, i, comparisons, end_time - start_time
    end_time = time.time()
    return False, -1, comparisons, end_time - start_time
def binary_search(arr, target):# Binary Search Algorithm
    comparisons = 0
    start_time = time.time()
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        if arr[mid] == target:
            end_time = time.time()
            return True, mid, comparisons, end_time - start_time
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    end_time = time.time()
    return False, -1, comparisons, end_time - start_time
# Input product list
n = int(input("Enter the number of product IDs: "))
product_ids = list(map(int, input("Enter the product IDs separated by space: ").split()))
# Input product ID to search
target = int(input("Enter the product ID to search: "))
# Linear Search
print("\n--- Linear Search ---")
found, index, comparisons, duration = linear_search(product_ids, target)
if found:
    print(f"Product found at index {index}")
else:
    print("Product not found")
print(f"Comparisons made: {comparisons}")
print(f"Time taken: {duration:.6f} seconds")
# Binary Search (needs sorted list)
sorted_ids = sorted(product_ids)
print("\n--- Binary Search (after sorting) ---")
found, index, comparisons, duration = binary_search(sorted_ids, target)
if found:
    print(f"Product found at index {index} in the sorted list")
else:
    print("Product not found in sorted list")
print(f"Comparisons made: {comparisons}")
print(f"Time taken: {duration:.6f} seconds")
```

IV. Given an array of integers nums and an integer target, return indices of the two numbers such
that they add up to target.
You may assume that each input would have exactly one solution, and you may not use
the same element twice.
You can return the answer in any order.

```
def two_sum(nums, target):
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    return []
def main():#Main function to demonstrate the two_sum solution with example inputs.
    print("--- Two Sum Problem Solver ---")
    # Example 1
    nums1 = [2, 7, 11, 15]
    target1 = 9
    print(f"\nInput: nums = {nums1}, target = {target1}")
    result1 = two_sum(nums1, target1)
    print(f"Output: {result1}") # Expected: [0, 1] or [1, 0]
    # Example 2
    nums2 = [3, 2, 4]
    target2 = 6
    print(f"\nInput: nums = {nums2}, target = {target2}")
    result2 = two_sum(nums2, target2)
    print(f"Output: {result2}") # Expected: [1, 2] or [2, 1]
if __name__ == "__main__":
    main()
```
