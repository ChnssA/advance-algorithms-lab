time complexity classes:
1. Constant:             
2. Logarithmic:           
3. Linear:

6. Cubic:               
7. Exponential:           
8. Factorial:

f(n) = n × log₂(n)

For better visualization, choose n from 1 to 20. Use logarithmic scaling or limit the y-axis range
where appropriate to accommodate fast-growing functions like 2ⁿ and n!.

```
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the range for n
n_values = np.arange(1, 21) # n from 1 to 20

# --- Define the functions for each time complexity class ---

# 1. Constant: f(n) = 1
def constant_fn(n):
    return np.ones_like(n, dtype=float) # Returns an array of 1s with the same shape as n

# 2. Logarithmic: f(n) = log₂(n)
def logarithmic_fn(n):
    # Handle log(0) if n could be 0, but our n starts from 1, so it's safe.
    return np.log2(n)

# 3. Linear: f(n) = n
def linear_fn(n):
    return n

# 4. Linearithmic: f(n) = n × log₂(n)
def linearithmic_fn(n):
    # For n=1, log2(1) is 0, so 1 * 0 = 0. This is correct.
    return n * np.log2(n)

# 5. Quadratic: f(n) = n²
def quadratic_fn(n):
    return n**2

# 6. Cubic: f(n) = n³
def cubic_fn(n):
    return n**3

# 7. Exponential: f(n) = 2ⁿ
def exponential_fn(n):
    return 2**n

# 8. Factorial: f(n) = n!
def factorial_fn(n):
    # Use math.factorial for individual values, then convert to numpy array
    return np.array([math.factorial(int(val)) for val in n])

# --- Calculate function values for n_values ---
y_constant = constant_fn(n_values)
y_logarithmic = logarithmic_fn(n_values)
y_linear = linear_fn(n_values)
y_linearithmic = linearithmic_fn(n_values)
y_quadratic = quadratic_fn(n_values)
y_cubic = cubic_fn(n_values)
y_exponential = exponential_fn(n_values)
y_factorial = factorial_fn(n_values)

# --- Plotting ---

# Create two subplots for better visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle('Time Complexity Classes: Runtime Growth with Input Size (n)', fontsize=16)

# Plot 1: Slower-growing functions (Linear Y-axis)
ax1.plot(n_values, y_constant, label='O(1) - Constant', marker='o', markersize=4)
ax1.plot(n_values, y_logarithmic, label='O(log n) - Logarithmic', marker='o', markersize=4)
ax1.plot(n_values, y_linear, label='O(n) - Linear', marker='o', markersize=4)
ax1.plot(n_values, y_linearithmic, label='O(n log n) - Linearithmic', marker='o', markersize=4)
ax1.plot(n_values, y_quadratic, label='O(n²) - Quadratic', marker='o', markersize=4)
ax1.plot(n_values, y_cubic, label='O(n³) - Cubic', marker='o', markersize=4)

ax1.set_title('Common Complexity Classes (Linear Scale)')
ax1.set_xlabel('Input Size (n)')
ax1.set_ylabel('Operations/Time')
ax1.legend(loc='upper left', bbox_to_anchor=(1, 1)) # Place legend outside the plot
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_xticks(n_values) # Ensure all n values are marked on x-axis
ax1.set_ylim(0, max(y_cubic) * 1.1) # Limit y-axis for better visibility of lower curves

# Plot 2: Faster-growing functions (Logarithmic Y-axis)
# Note: Factorial values grow extremely fast. For n=20, 20! is huge.
# Logarithmic scale helps visualize the difference.
ax2.plot(n_values, y_exponential, label='O(2ⁿ) - Exponential', marker='o', markersize=4, color='red')
ax2.plot(n_values, y_factorial, label='O(n!) - Factorial', marker='o', markersize=4, color='purple')

ax2.set_title('High Complexity Classes (Logarithmic Scale)')
ax2.set_xlabel('Input Size (n)')
ax2.set_ylabel('Operations/Time (Log Scale)')
ax2.set_yscale('log') # Set y-axis to logarithmic scale
ax2.legend(loc='upper left', bbox_to_anchor=(1, 1)) # Place legend outside the plot
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_xticks(n_values) # Ensure all n values are marked on x-axis

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0.03, 0.85, 0.95]) # Adjust rect to make space for suptitle and legends
plt.show()
```

. A teacher is preparing the final merit list for a class test. Each student has a name and a
score out of 100. The list must be sorted so that:
• Students are arranged in descending order of marks.
• If two students have the same mark, they should be sorted alphabetically by name.
Write a program that:
1. Accepts input for n students: name (string) and marks (integer).
2. Sorts the student records using a custom implementation of a sorting algorithm
3. Displays the sorted list in a clean, tabular format showing rank, name, and marks.
4. Tracks and displays:
o Total number of comparisons
o Total number of swaps

```
import math

def custom_bubble_sort(students):
    """
    Sorts a list of student dictionaries using a custom Bubble Sort algorithm.
    Primary sort key: marks (descending).
    Secondary sort key: name (alphabetical, ascending) for ties in marks.

    Args:
        students (list): A list of dictionaries, where each dictionary
                         represents a student with 'name' (str) and 'marks' (int).

    Returns:
        tuple: A tuple containing the sorted list of students, total comparisons,
               and total swaps.
    """
    n = len(students)
    total_comparisons = 0
    total_swaps = 0

    # Outer loop for passes
    for i in range(n - 1):
        swapped = False # Flag to optimize: if no swaps in a pass, list is sorted
        # Inner loop for comparisons and swaps
        for j in range(0, n - i - 1):
            total_comparisons += 1 # Increment comparison counter for each pair checked

            student_a = students[j]
            student_b = students[j + 1]

            # Custom comparison logic:
            # Sort by marks in descending order first
            # If marks are equal, sort by name in ascending (alphabetical) order

            should_swap = False
            if student_a['marks'] < student_b['marks']:
                # If current student has fewer marks than the next, they are in wrong order (descending marks)
                should_swap = True
            elif student_a['marks'] == student_b['marks']:
                # If marks are equal, compare names alphabetically
                if student_a['name'] > student_b['name']:
                    # If current student's name comes after next student's name, they are in wrong order (alphabetical)
                    should_swap = True

            if should_swap:
                students[j], students[j + 1] = students[j + 1], students[j]
                total_swaps += 1 # Increment swap counter
                swapped = True

        # If no two elements were swapped by inner loop, then break
        if not swapped:
            break

    return students, total_comparisons, total_swaps

def get_student_input():
    """
    Accepts student name and marks input from the user.

    Returns:
        list: A list of student dictionaries.
    """
    students = []
    while True:
        try:
            num_students = int(input("Enter the number of students (e.g., 5): "))
            if num_students <= 0:
                print("Please enter a positive number of students.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    print("\nEnter student details:")
    for i in range(num_students):
        while True:
            name = input(f"Enter name for student {i+1}: ").strip()
            if name: # Ensure name is not empty
                break
            else:
                print("Name cannot be empty. Please try again.")

        while True:
            try:
                marks = int(input(f"Enter marks for {name} (0-100): "))
                if 0 <= marks <= 100:
                    break
                else:
                    print("Marks must be between 0 and 100. Please try again.")
            except ValueError:
                print("Invalid input. Please enter an integer for marks.")

        students.append({'name': name, 'marks': marks})
    return students

def display_merit_list(students):
    """
    Displays the sorted student list in a tabular format with ranks.

    Args:
        students (list): The list of sorted student dictionaries.
    """
    print("\n--- Final Merit List ---")
    print("{:<5} {:<20} {:<5}".format("Rank", "Name", "Marks"))
    print("-" * 35)

    for i, student in enumerate(students):
        print("{:<5} {:<20} {:<5}".format(i + 1, student['name'], student['marks']))
    print("-" * 35)

def main():
    """
    Main function to run the student sorting program.
    """
    students_data = get_student_input()

    if not students_data:
        print("No student data entered. Exiting.")
        return

    print("\nSorting student data...")
    sorted_students, comparisons, swaps = custom_bubble_sort(students_data)

    display_merit_list(sorted_students)

    print(f"\nSorting Metrics:")
    print(f"Total comparisons: {comparisons}")
    print(f"Total swaps: {swaps}")

if __name__ == "__main__":
    main()
```

III. You are developing a module for a retail store’s inventory management system. The system
stores a list of product IDs (integers) and needs to quickly locate a product when a customer
requests it.
1. Accept a list of n product IDs (integers) as input.
2. Accept a product ID to search (target).
3. Implement the search algorithms
4. For each algorithm:
o Return whether the product is found and at what index.
o Count and display the number of comparisons made.
o Measure the time taken to perform the search.

In an actual software development scenario, especially for a retail store's inventory management system, it's crucial to consider that product IDs might not always be sorted. Therefore, both linear and binary search algorithms have their place. A linear search would be applicable for unsorted lists, while a binary search would be highly efficient for sorted lists. The choice of algorithm would depend on whether the product ID list is guaranteed to be sorted or if sorting is feasible as a preprocessing step.

Let's outline the plan for developing the module:

Input Collection:

Prompt the user to enter the number of product IDs (n).

Prompt the user to enter n product IDs, ensuring they are integers.

Prompt the user to enter the target product ID to search for.

Algorithm Implementations:

Linear Search:

Iterate through the list from the beginning to the end.

Compare each element with the target.

Increment a comparison counter for each comparison.

If the target is found, return True and its index.

If the end of the list is reached without finding the target, return False.

Binary Search (requires a sorted list):

First, ensure the input list of product IDs is sorted. This is a critical prerequisite for binary search. We'll sort it if it's not already.

Initialize low to 0 and high to n-1.

While low is less than or equal to high:

Calculate mid = (low + high) // 2.

Increment a comparison counter.

If list[mid] is equal to target, return True and mid.

If list[mid] is less than target, set low = mid + 1.

If list[mid] is greater than target, set high = mid - 1.

If the loop finishes, return False.

Performance Measurement and Display:

For each algorithm:

Record the start time before executing the search.

Execute the search algorithm.

Record the end time after the search.

Calculate the time taken (end_time - start_time).

Display:

Whether the product was found and its index (if found).

The total number of comparisons made.

The time taken for the search (e.g., in milliseconds or microseconds).

This comprehensive approach will demonstrate the functionality and performance characteristics of both search algorithms, which is crucial for making informed decisions in an inventory management system.

```
import time
import sys

# Set a higher recursion limit for binary search, though iterative is preferred for large lists
sys.setrecursionlimit(2000)

def get_product_ids():
    """Accepts a list of n product IDs (integers) as input."""
    while True:
        try:
            n = int(input("Enter the number of product IDs: "))
            if n <= 0:
                print("Number of product IDs must be positive. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    product_ids = []
    print(f"Enter {n} product IDs:")
    for i in range(n):
        while True:
            try:
                product_id = int(input(f"Product ID {i+1}: "))
                product_ids.append(product_id)
                break
            except ValueError:
                print("Invalid input. Please enter an integer.")
    return product_ids

def get_target_id():
    """Accepts a product ID to search (target)."""
    while True:
        try:
            target = int(input("Enter the product ID to search for: "))
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")
    return target

def linear_search(product_list, target):
    """
    Implements the linear search algorithm.
    Returns (found, index, comparisons, time_taken)
    """
    comparisons = 0
    start_time = time.perf_counter_ns()

    for i in range(len(product_list)):
        comparisons += 1
        if product_list[i] == target:
            end_time = time.perf_counter_ns()
            return True, i, comparisons, (end_time - start_time)

    end_time = time.perf_counter_ns()
    return False, -1, comparisons, (end_time - start_time)

def binary_search_iterative(sorted_product_list, target):
    """
    Implements the iterative binary search algorithm.
    Requires a sorted list.
    Returns (found, index, comparisons, time_taken)
    """
    comparisons = 0
    start_time = time.perf_counter_ns()

    low = 0
    high = len(sorted_product_list) - 1

    while low <= high:
        mid = (low + high) // 2
        comparisons += 1

        if sorted_product_list[mid] == target:
            end_time = time.perf_counter_ns()
            return True, mid, comparisons, (end_time - start_time)
        elif sorted_product_list[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    end_time = time.perf_counter_ns()
    return False, -1, comparisons, (end_time - start_time)

def binary_search_recursive_helper(sorted_product_list, target, low, high, comparisons_ref):
    """Helper for recursive binary search to track comparisons."""
    if low > high:
        return False, -1

    comparisons_ref[0] += 1
    mid = (low + high) // 2

    if sorted_product_list[mid] == target:
        return True, mid
    elif sorted_product_list[mid] < target:
        return binary_search_recursive_helper(sorted_product_list, target, mid + 1, high, comparisons_ref)
    else:
        return binary_search_recursive_helper(sorted_product_list, target, low, mid - 1, comparisons_ref)

def binary_search_recursive(sorted_product_list, target):
    """
    Implements the recursive binary search algorithm.
    Requires a sorted list.
    Returns (found, index, comparisons, time_taken)
    """
    comparisons = [0]  # Use a list to pass comparisons by reference
    start_time = time.perf_counter_ns()

    found, index = binary_search_recursive_helper(sorted_product_list, target, 0, len(sorted_product_list) - 1, comparisons)

    end_time = time.perf_counter_ns()
    return found, index, comparisons[0], (end_time - start_time)


def display_results(algorithm_name, found, index, comparisons, time_taken_ns):
    """Displays the results for a given search algorithm."""
    print(f"\n--- {algorithm_name} Results ---")
    if found:
        print(f"Product found at index: {index}")
    else:
        print("Product not found.")
    print(f"Number of comparisons made: {comparisons}")
    print(f"Time taken: {time_taken_ns / 1_000_000:.6f} milliseconds") # Convert nanoseconds to milliseconds

def main():
    product_ids = get_product_ids()
    target_id = get_target_id()

    # --- Linear Search ---
    found_linear, index_linear, comparisons_linear, time_linear = linear_search(product_ids, target_id)
    display_results("Linear Search", found_linear, index_linear, comparisons_linear, time_linear)

    # --- Binary Search (Iterative) ---
    # Binary search requires a sorted list
    sorted_product_ids = sorted(product_ids)
    print("\n--- Note: For Binary Search, the list is sorted ---")
    print(f"Sorted Product IDs: {sorted_product_ids}")

    found_binary_iterative, index_binary_iterative, comparisons_binary_iterative, time_binary_iterative = \
        binary_search_iterative(sorted_product_ids, target_id)
    display_results("Binary Search (Iterative)", found_binary_iterative, index_binary_iterative, comparisons_binary_iterative, time_binary_iterative)

    # --- Binary Search (Recursive) ---
    found_binary_recursive, index_binary_recursive, comparisons_binary_recursive, time_binary_recursive = \
        binary_search_recursive(sorted_product_ids, target_id)
    display_results("Binary Search (Recursive)", found_binary_recursive, index_binary_recursive, comparisons_binary_recursive, time_binary_recursive)

if __name__ == "__main__":
    main()
```

```
import time

# Linear Search Algorithm
def linear_search(arr, target):
    comparisons = 0
    start_time = time.time()

    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            end_time = time.time()
            return True, i, comparisons, end_time - start_time

    end_time = time.time()
    return False, -1, comparisons, end_time - start_time

# Binary Search Algorithm
def binary_search(arr, target):
    comparisons = 0
    start_time = time.time()

    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        if arr[mid] == target:
            end_time = time.time()
            return True, mid, comparisons, end_time - start_time
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    end_time = time.time()
    return False, -1, comparisons, end_time - start_time

# Input product list
n = int(input("Enter the number of product IDs: "))
product_ids = list(map(int, input("Enter the product IDs separated by space: ").split()))

# Input product ID to search
target = int(input("Enter the product ID to search: "))

# Linear Search
print("\n--- Linear Search ---")
found, index, comparisons, duration = linear_search(product_ids, target)
if found:
    print(f"Product found at index {index}")
else:
    print("Product not found")
print(f"Comparisons made: {comparisons}")
print(f"Time taken: {duration:.6f} seconds")

# Binary Search (needs sorted list)
sorted_ids = sorted(product_ids)

print("\n--- Binary Search (after sorting) ---")
found, index, comparisons, duration = binary_search(sorted_ids, target)
if found:
    print(f"Product found at index {index} in the sorted list")
else:
    print("Product not found in sorted list")
print(f"Comparisons made: {comparisons}")
print(f"Time taken: {duration:.6f} seconds")
```

```
import time

def linear_search(arr, target):
    comparisons = 0
    start_time = time.perf_counter()
    for idx, item in enumerate(arr):
        comparisons += 1
        if item == target:
            time_taken = time.perf_counter() - start_time
            return True, idx, comparisons, time_taken
    time_taken = time.perf_counter() - start_time
    return False, -1, comparisons, time_taken

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    comparisons = 0
    start_time = time.perf_counter()
    while left <= right:
        mid = (left + right) // 2
        comparisons += 1
        if arr[mid] == target:
            time_taken = time.perf_counter() - start_time
            return True, mid, comparisons, time_taken
        comparisons += 1
        if arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    time_taken = time.perf_counter() - start_time
    return False, -1, comparisons, time_taken

def main():
    n = int(input("Enter number of products: "))
    print("Enter product IDs (integers):")
    product_ids = [int(input(f"ID #{i+1}: ")) for i in range(n)]

    target = int(input("\nEnter product ID to search for: "))

    print("\n1. Linear Search")
    found, idx, comps, t = linear_search(product_ids, target)
    if found:
        print(f"Product found at index {idx}")
    else:
        print("Product not found")
    print(f"Comparisons: {comps}, Time taken: {t:.6f} seconds")

    # Binary search requires sorted list
    sorted_ids = sorted(product_ids)
    print("\n2. Binary Search (on sorted list)")
    found, idx, comps, t = binary_search(sorted_ids, target)
    if found:
        print(f"Product found at index {idx} (in sorted list)")
    else:
        print("Product not found")
    print(f"Comparisons: {comps}, Time taken: {t:.6f} seconds")
    print("\n(Note: Indices may be different due to sorting in binary search)")


if __name__ == '__main__':
    main()
```

```
def two_sum(nums, target):
    """
    Given an array of integers nums and an integer target,
    return indices of the two numbers such that they add up to target.

    You may assume that each input would have exactly one solution,
    and you may not use the same element twice.
    You can return the answer in any order.

    Args:
        nums (list[int]): A list of integers.
        target (int): The target sum.

    Returns:
        list[int]: A list containing the indices of the two numbers that sum to the target.
                   Returns an empty list if no solution is found (though the problem states
                   there will always be exactly one solution).
    """
    # Create a dictionary to store numbers and their indices.
    # This allows for O(1) average time complexity lookup.
    num_map = {}

    # Iterate through the list with both index and value
    for i, num in enumerate(nums):
        # Calculate the complement needed to reach the target
        complement = target - num

        # Check if the complement already exists in our map
        # If it does, we've found our two numbers
        if complement in num_map:
            # Return the index of the complement and the current index
            return [num_map[complement], i]

        # If the complement is not found, add the current number and its index to the map
        num_map[num] = i

    # According to the problem statement, there will always be exactly one solution,
    # so this line should ideally not be reached.
    return []

def main():
    """
    Main function to demonstrate the two_sum solution with example inputs.
    """
    print("--- Two Sum Problem Solver ---")

    # Example 1
    nums1 = [2, 7, 11, 15]
    target1 = 9
    print(f"\nInput: nums = {nums1}, target = {target1}")
    result1 = two_sum(nums1, target1)
    print(f"Output: {result1}") # Expected: [0, 1] or [1, 0]

    # Example 2
    nums2 = [3, 2, 4]
    target2 = 6
    print(f"\nInput: nums = {nums2}, target = {target2}")
    result2 = two_sum(nums2, target2)
    print(f"Output: {result2}") # Expected: [1, 2] or [2, 1]

    # Example 3
    nums3 = [3, 3]
    target3 = 6
    print(f"\nInput: nums = {nums3}, target = {target3}")
    result3 = two_sum(nums3, target3)
    print(f"Output: {result3}") # Expected: [0, 1] or [1, 0]

    # Example 4: A larger list
    nums4 = [1, 5, 9, 10, 20, 30, 40, 50]
    target4 = 70
    print(f"\nInput: nums = {nums4}, target = {target4}")
    result4 = two_sum(nums4, target4)
    print(f"Output: {result4}") # Expected: [5, 7] or [7, 5]

if __name__ == "__main__":
    main()
```
